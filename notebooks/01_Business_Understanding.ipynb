## 1. O que é o CRISP-DM?

O **CRISP-DM** (*Cross-Industry Standard Process for Data Mining*) é uma metodologia amplamente utilizada em projetos de análise de dados e machine learning. Organiza o trabalho em fases sequenciais, tornando o processo mais estruturado e reproduzível.

As fases são:

1. **Business Understanding** — compreender o problema de negócio e o que se pretende alcançar
2. **Data Understanding** — explorar os dados disponíveis e perceber a sua estrutura
3. **Data Preparation** — limpar e preparar os dados para análise
4. **Modelling** — construir modelos preditivos
5. **Evaluation** — avaliar se os modelos respondem ao objetivo inicial
6. **Deployment** — colocar em produção (fora do âmbito deste projeto)

## 6. Métricas de Avaliação

O dataset deste projecto tem classes desbalanceadas: a maioria dos candidatos não procura mudar de emprego, e apenas uma minoria procura. Nestes casos, a Accuracy deixa de ser uma boa métrica, porque um modelo que classifique toda a gente como "não muda" ainda pode ter Accuracy elevada. Por isso utilizamos as seguintes métricas:

**Precision** mede, de todos os candidatos que o modelo classificou como prováveis a mudar de emprego, quantos realmente queriam mudar. Uma Precision elevada significa que o modelo é criterioso: quando sinaliza alguém, geralmente acerta. Em contexto de RH, isto evita desperdiçar esforço de retenção com pessoas que não tinham intenção de sair.

**Recall** mede, de todos os candidatos que realmente queriam mudar de emprego, quantos o modelo conseguiu identificar. Um Recall elevado significa que o modelo é sensível: não deixa escapar muitos casos reais. Em contexto de RH, falhar a identificação de um colaborador que está prestes a sair tem um custo elevado em tempo de recrutamento e perda de conhecimento.

**F1-Score** é a média harmónica entre Precision e Recall. Quando existe um desequilíbrio entre classes, o F1 é preferível à Accuracy porque pondera os dois tipos de erro de forma equilibrada. É a métrica principal deste projecto.

**AUC (Area Under the ROC Curve)** mede a capacidade do modelo de distinguir entre as duas classes, independentemente do limiar de decisão escolhido. Um AUC próximo de 1 indica que o modelo separa bem os casos positivos dos negativos. É útil para comparar modelos entre si e para avaliar o poder discriminatório mesmo quando as classes são muito desiguais em tamanho.

## 7. Critérios de Sucesso

Para que o modelo seja considerado útil neste projecto, os critérios de desempenho mínimos são os seguintes:

- **F1-Score igual ou superior a 0.70**, garantindo um equilíbrio aceitável entre Precision e Recall no contexto de classes desbalanceadas.
- **AUC igual ou superior a 0.80**, demonstrando que o modelo tem capacidade real de distinguir candidatos que mudam de emprego dos que ficam.
- **Recall igual ou superior a 0.65**, assegurando que o modelo identifica a maioria dos casos positivos reais, o que é prioritário em contextos de retenção de talento.

Para além do desempenho quantitativo, espera-se que o projecto permita:

- Identificar as variáveis que mais influenciam a decisão de mudar de emprego.
- Produzir uma análise interpretável dos resultados, útil para as equipas de Recursos Humanos.
- Fornecer informação accionável que apoie decisões de retenção com base em dados e não apenas em intuição.

Um modelo que cumpra estes critérios será considerado adequado para apoio à decisão em contexto real.
